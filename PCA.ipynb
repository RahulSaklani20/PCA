```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
```

```python
wine=pd.read_csv('wine.csv')
wine
```

```python
wine.describe()
```

```python
wine.info()
```

```python
wine.shape
```

```python
#coverting data to numpy array
wine_ar=wine.values
wine_ar
```

```python
#Normalizing the Numerical Data
wine_nor=scale(wine_ar)
wine_nor
```

#### PCA Implementaytion

```python
pca=PCA()
pca_val=pca.fit_transform(wine_nor)
pca_val
```

```python
pca.components_
```

```python
# The amount of variance that each PCA has
var=pca.explained_variance_ratio_
var
```

```python
# Cummulative variance of each PCA
var1=np.cumsum(np.round(var,decimals=4)*100)
var1
```

```python
plt.plot(var1)
```

```python
#final Dataframe
final_df=pd.concat([wine['Type'],pd.DataFrame(pca_val[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
final_df
```

```python
#Visualization
fig=plt.figure(figsize=(20,15))
sns.scatterplot(data=final_df)
```

```python
sns.scatterplot(data=final_df,x='PC1',y='PC2',hue='Type');
```

# By Using Clustering Algorithms

### Hierarchical Clustering

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
plt.figure(figsize=(15,12))
dendrogram=sch.dendrogram(sch.linkage(wine_nor,'complete'))
```

```python
hcluster=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hcluster
```

```python
y=pd.DataFrame(hcluster.fit_predict(wine_nor),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
wine2=wine.copy()
```

```python
wine2['clustersid']=hcluster.labels_
wine2
```

### KMean Clustering

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,6):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(wine_nor)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,6),wcss)
plt.title('Elbow Graph')
plt.xlabel('No. Of Clusters')
plt.ylabel('WCSS')
```

### Building Cluster Using K=3

```python
cluster3=KMeans(3,random_state=30).fit(wine_nor)
cluster3
```

```python
cluster3.labels_
```

```python
wine3=wine.copy()
```

```python
wine3['clusters3id']=cluster3.labels_
wine3
```

```python
wine3['clusters3id'].value_counts()
```
